```python
%%writefile app.py
import streamlit as st
from streamlit_drawable_canvas import st_canvas
from tensorflow import keras
import numpy as np
import cv2

model = keras.models.load_model('/content/drive/MyDrive/minor project/mnist.hdf5')

st.title('Digit Recognizer')
st.markdown('''
Try to write a digit!
''')

st.title('orginal image')

SIZE = 192
# Create a canvas component
canvas_result = st_canvas(fill_color='#FFFFFF',stroke_width=10,stroke_color='#000000',height=250,width =250,drawing_mode='freedraw',)

if canvas_result.image_data is not None:
    st.image(canvas_result.image_data)

if canvas_result.image_data is not None:
    img = cv2.cvtColor(canvas_result.image_data.astype('uint8'), cv2.COLOR_BGR2GRAY)
    rescaled = cv2.resize(img, (28,28))
    
if st.button('Predict'):
    op = model.predict(rescaled.reshape(1,28,28))
    st.title(f'result: {np.argmax(op[0])}')
    st.bar_chart(op[0])

```
